```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [3,5,0,6,2,5]
plt.plot(x,y)
plt.show()
```

# Line Graph

```python
x = [1,2,3,4,5,6]
y = [3,5,0,6,2,5]
plt.plot(x,y)
plt.title("Basic Line Graph")
plt.xlabel("X label here")
plt.ylabel("Y label here")
plt.show()
```

# Bar chart

```python
x = [1,2,8,3,4,5,6]
y = [4,7,3,5,6,1,4]
plt.bar(x,y)
plt.show()
```

# Histogram

```python
x = [14,25,83,3,3,7,12,25,26,25,11,12,7,7,8,9,5,3,2,34,40,39,18,28,24,13.,17,19,22,21,20]
y = [0,5,10,15,20,25,30,35,40]
plt.hist(x,y,edgecolor='black')
plt.show()
```

# scaller plot


```python
x = [5,8,9,3,4,2,6,7,6,0,1,8,2]
y = [4,2,6,8,3,5,6,9,0,1,3,8,6]
plt.scatter(x,y)
plt.show()
```

# stack plot

```python
#If there is more y value for x, we use this.

day = [1,2,3,4,5]

google = [50, 60, 55, 45, 65]
apple = [45, 55, 65, 35, 38]
facebook = [65, 33, 45, 28, 62]
microsoft = [55, 60, 67, 72, 55]

plt.stackplot(day, google, apple, facebook, microsoft)
plt.show()
```
